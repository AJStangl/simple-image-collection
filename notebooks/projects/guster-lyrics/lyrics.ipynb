```python
import pandas
import json
import re

songs = []

with open("Lyrics_Guster.json", 'r', encoding='utf=8') as f:
    data = f.read()
    json_data = json.loads(data)
    songs = json_data['songs']


df = pandas.DataFrame(data=songs)
more_data = []
records = df.to_dict(orient='records')

for elem in records:
    try:
        al_data = elem['album']['name']
    except:
        al_data = ""
    foo = {'title': elem['title'], 'lyrics': elem['lyrics'], 'album': al_data }
    more_data.append(foo)

df = pandas.DataFrame(data=more_data)


def parse_lyrics(x: object) -> str:
    song = x['lyrics']
    album = x['album']
    out_line = f"<|startoftext|><|album|>{album}"
    song = song.replace("[", "<|")
    song = song.replace("]", "|>")
    final = out_line + song + "<|endoftext|>"
    final = final.replace("\n", "")
    final = final.replace("|>\\n", "|>")
    final = final.replace("\\n<|", "<|")
    final += "\n"
    return final

df['training'] = df.apply(lambda x: parse_lyrics(x), axis=1)

display(df)

# records = df.to_dict(orient='records')
#
# tks = []
# for record in records:
#     training = record['training']
#     foos = re.findall("\<\|(.+?)\|\>", training)
#     for item in foos:
#         tks.append(item)

# r = list(set(tks))
# for elem in r:
#     print(elem)
# # with open('foo.txt', 'wb') as b:
# #     for record in records:
# #         b.write(record['training'].encode('utf-8'))
```
